```python
import pandas as pd
df = pd.read_csv('dor.csv')
df
```

```python
df.describe
```

```python
df.info()
```

```python
df['year'] = pd.to_numeric(df['year'])
```

```python
pd.set_option('display.max_rows' ,500)
pd.set_option('display.max_columns' ,500)
pd.set_option('display.width',1000)
```

## Oosc rates of all the states in the year range of 2012-15

```python
import matplotlib.pyplot as plt
plt.plot(df['year'],df['State_UT'],label='dor')
plt.legend()
```

## Oosc rate of primary boys over the states

```python
plt.plot(df['State_UT'],df['Primary_Boys'],label ='pri_boys')
plt.grid(True)
plt.legend()
```

## Oosc rates of primary boys over the years

```python
plt.plot(df['year'],df['Primary_Boys'],label ='pri_boys')
plt.grid(True)
plt.legend()
```

## Histogram
## Oosc rates of primary students over the states in the year range of 2012-15

```python
fig,axes= plt.subplots(2,2)
axes[0,0].hist(df['State_UT'])
axes[0,1].plot(df['year'],df['Primary_Boys'])
axes[1,0].plot(df['year'],df['Primary_Girls'])
axes[1,1].plot(df['year'],df['Primary_Total'])
```

## Out-of-school students identification over the years 2012-15

```python
plt.plot(df['year'],df['Primary_Total'],label ='Primary_Total')
plt.plot(df['year'],df['Upper Primary_Total'],label ='Upper Primary_Total')
plt.plot(df['year'],df['Secondary _Total'],label ='Secondary _Total')
plt.plot(df['year'],df['HrSecondary_Total'],label ='HrSecondary_Total')
plt.grid(True)
plt.legend()
```

```python
plt.plot(df['year'],df['Primary_Total'],'r-')
plt.plot(df['year'],df['Upper Primary_Total'],'bs')
plt.plot(df['year'],df['Secondary _Total'], 'g^')
plt.plot(df['year'],df['HrSecondary_Total'],'p-')
```

## Scatterplot

```python
plt.scatter(df['year'],df['Primary_Total'])
plt.scatter(df['year'],df['Upper Primary_Total'])
plt.scatter(df['year'],df['Secondary _Total'],)
plt.scatter(df['year'],df['HrSecondary_Total'])
```

```python
import seaborn as sns
sns.set()
sns.scatterplot(df['Primary_Total'])
```
